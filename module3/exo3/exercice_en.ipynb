# Covid-19

## I. Analysis number of cases covid-19

### 1.1 Import raw data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('https://app-learninglab.inria.fr/moocrr/gitlab/082aa0aa9507b80b621099010e33de9d/mooc-rr/raw/master/module3/exo3/time_series_covid19_confirmed_global.csv')
df
```

### 1.2 Sorting by chosen country

```python
# Country list
countries_of_interest = ["Belgium", "China", "Hong Kong", "France", "Germany", "Iran", "Italy", "Japan",
                         "Korea, South", "Netherlands", "Portugal", "Spain", "United Kingdom", "US"]

# Filter by country
filtered_data = df[df['Country/Region'].isin(countries_of_interest)]

filtered_data
```

### 1.3 Remove no-needed data

```python
# Filter country and value
filtered_data = filtered_data.drop(columns=['Province/State', 'Lat', 'Long'])
filtered_data
```

### 1.4 Data in chosen country

```python
# Group by country and sum cases for each date
grouped_data = filtered_data.groupby('Country/Region').sum()
grouped_data
```

### 1.5 Plot the graph

#### 1.5.1 Graph linear day by day

```python
# Select date columns
date_columns = grouped_data.columns  # All remaining columns should be date columns

# Plot the data
plt.figure(figsize=(14, 8))

for country in countries_of_interest:
    if country in grouped_data.index:
        plt.plot(date_columns, grouped_data.loc[country], label=country)

plt.xlabel('Date')
plt.ylabel('Cumulative Number of Cases')
plt.title('Cumulative Number of COVID-19 Cases Over Time - Linear - day view')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### 1.5.2 Graph linear with better x-axis visualization (by months)

```python
date_columns = pd.to_datetime(date_columns)
# Plot the data
plt.figure(figsize=(14, 8))

for country in countries_of_interest:
    if country in grouped_data.index:
        plt.plot(date_columns, grouped_data.loc[country], label=country)

plt.xlabel('Date')
plt.ylabel('Cumulative Number of Cases')
plt.title('Cumulative Number of COVID-19 Cases Over Time - Linear - Month view')

# Format x-axis to show month and year
# Sets the major formatter for the x-axis to display the month and year.
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
# Sets the major locator to display ticks for each month.
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator(interval=1))

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### 1.5.3 Graph logarit with better x-axis visualization (by months)

```python
# Convert date columns to datetime for better plotting
date_columns = pd.to_datetime(date_columns)

# Plot the data
plt.figure(figsize=(14, 8))

for country in countries_of_interest:
    if country in grouped_data.index:
        plt.plot(date_columns, grouped_data.loc[country], label=country)

plt.xlabel('Date')
plt.ylabel('Cumulative Number of Cases')
plt.title('Cumulative Number of COVID-19 Cases Over Time - Loga - Month view')

# Format x-axis to show month and year
# Sets the major formatter for the x-axis to display the month and year.
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
# Sets the major locator to display ticks for each month.
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator(interval=1))

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.yscale('log')
plt.show()
```

## III. France daily number

### 3.1 Import data

```python
monthly_data = pd.read_csv('https://app-learninglab.inria.fr/moocrr/gitlab/082aa0aa9507b80b621099010e33de9d/mooc-rr/raw/master/module3/exo3/monthly_values_1.csv?inline=false')
monthly_data
```

### 3.2 Cumumlative sum data by month

```python

# Exclude the first column (Label) and convert the remaining columns to numeric format (assuming they represent numbers)
data_columns = monthly_data.columns[1:]
monthly_data[data_columns] = monthly_data[data_columns].apply(pd.to_numeric)

# Calculate the cumulative sum for each month
new_cul_data = monthly_data.copy()  # Create a copy to avoid modifying the original DataFrame
new_cul_data[data_columns] = new_cul_data[data_columns].cumsum(axis=1)

# Display the new DataFrame with cumulative data
display(new_cul_data)
```

### 3.2 Plot the graph

```python
# Set the date range for the x-axis labels
date_range = new_cul_data.columns[1:]

# Plot the cumulative data
plt.figure(figsize=(14, 8))
for index, row in new_cul_data.iterrows():
    plt.plot(date_range, row[1:], label=row['Label'])

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Cumulative Number')
plt.title('Cumulative Number of Deaths Over Time')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
